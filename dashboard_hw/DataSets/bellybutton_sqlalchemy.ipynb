```python
def calc_temps(start_date, end_date):
    start_date = start_date
    end_date = end_date
    
    sel = [Measurement.date,
           Measurement.tobs,
           func.avg(Measurement.tobs),
           func.max(Measurement.tobs),
           func.min(Measurement.tobs)]
    date_choice = session.query(*sel).\
        filter(Measurement.date >= start_date, Measurement.date <= end_date).\
        group_by(Measurement.date).\
        order_by(Measurement.date).all()
    
    return(date_choice)

calc_temps('2017-04-01','2017-04-12')
```

```python
def metadata(sample):
    sample = sample
    
    names4 = []
    result = session.query(Samples).all()
    for c in Samples.__table__.c:
        if c.name != "otu_id":
            names4.append(c.name)
#             print(c.name)
    
    sel = [Samples_metadata.AGE,
           Samples_metadata.BBTYPE,
           Samples_metadata.ETHNICITY,
           Samples_metadata.GENDER,
           Samples_metadata.LOCATION,
           Samples_metadata.SAMPLEID,
          ]

    results = session.query(*sel).\
        order_by(Samples_metadata.SAMPLEID).all()

    meta_samples = results

    keys = ["Age", "BBType", "Ethnicity", "Gender", "Location", "SampleID"]
    data = meta_samples

    meta_samples_dict = [dict(zip(keys, d)) for d in data]
    
    final_meta_dict = dict(zip(names4, meta_samples_dict))    
    
    for k, v in final_meta_dict.items():
        if k == sample:
            
            return(k, v)
    
metadata('BB_1233')    
    
    
    
    
    
    
    
    
    
    
    
```

```python
def washes(sample):
    sample = sample
    
    names5 = []
    result = session.query(Samples).all()
    for c in Samples.__table__.c:
        if c.name != "otu_id":
            names5.append(c.name)
#             print(c.name)
    
    sel = [Samples_metadata.WFREQ]

    results = session.query(*sel).all()

    washes = results

    keys = ["Wash Frequency"]
    data = washes

    washes_dict = [dict(zip(keys, d)) for d in data]

    wash_dict = dict(zip(names5, washes_dict))   
    
    for k, v in wash_dict.items():
        if k == sample:
            
            return(k, v)
    
washes('BB_1233')    
    
    
```
